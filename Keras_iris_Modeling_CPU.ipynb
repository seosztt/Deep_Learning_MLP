# Keras iris Modeling

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns

iris = sns.load_dataset('iris')
```

```python
iris.info()
```

```python
iris.head()
```

# Data Preprocessing

## iris.Species 빈도분석

```python
iris.species.value_counts()
```

##  DataFrame to Array & Casting

```python
iris_AR = iris.values

iris_AR
```

```python
AR_X = iris_AR[:, 0:4].astype(float)
AR_y = iris_AR[:, 4]

AR_X.shape, AR_y.shape
```

##  One Hot Encoding with sklearn & Keras

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
AR_yLBE = encoder.fit_transform(AR_y)

AR_yLBE
```

```python
from tensorflow.keras.utils import to_categorical

AR_yOHE = to_categorical(AR_yLBE)

AR_yOHE
```

## Version check

```python
import tensorflow

tensorflow.__version__
```

```python
import keras

keras.__version__
```

## train_test_split()

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(AR_X, AR_yOHE, 
                                                    test_size = 0.3, 
                                                    random_state = 2045) 

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Keras Modeling

## Keras models & layers Import

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

## Model Define

```python
Model_iris = models.Sequential()

Model_iris.add(layers.Dense(16, activation = 'relu', input_shape=(4, )))
Model_iris.add(layers.Dense(8, activation = 'relu'))
Model_iris.add(layers.Dense(3, activation = 'softmax'))
```

```python
Model_iris.summary()
```

```python
from tensorflow.keras import utils

utils.plot_model(Model_iris,
                 show_shapes = True,
                 show_dtype = True)
```

## Model Compile

```python
Model_iris.compile(loss='categorical_crossentropy',
                   optimizer='adam',
                   metrics = ['accuracy'])
```

## Model Fit

```python
History_iris = Model_iris.fit(X_train, y_train,
                              epochs = 500,
                              batch_size = 7,
                              validation_data = (X_test,y_test))
```

## 학습 결과 시각화

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))
plt.ylim(0, 1.2)
plt.plot(History_iris.history['loss'])
plt.plot(History_iris.history['val_loss'])
plt.plot(History_iris.history['accuracy'])
plt.plot(History_iris.history['val_accuracy'])
plt.legend(['loss','val_loss', 'accuracy','val_accuracy'])
plt.grid()
plt.show()
```

```python
import pandas as pd

pd.DataFrame(History_iris.history)
```

## Model Evaluate

```python
loss, accuracy = Model_iris.evaluate(X_test, y_test)

print('Loss = {:.2f}'.format(loss))
print('Accuracy = {:.2f}'.format(accuracy))
```

## Model Predict

```python
import numpy as np
np.set_printoptions(suppress = True, precision =5)

Model_iris.predict(X_test)
```

```python
y_hat=np.argmax(Model_iris.predict(X_test), axis=1)

y_hat
```

```python
y=np.argmax(y_test, axis=1)

y
```

```python
from sklearn.metrics import confusion_matrix, classification_report

confusion_matrix(y,y_hat)
```

```python
print(classification_report(y, y_hat,
                            target_names =['setosa',
                                           'virginica',
                                           'versicolor']))
```
